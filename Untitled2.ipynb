```python
def make_matching_figure(
        img0, img1, mkpts0, mkpts1, color,
        kpts0=None, kpts1=None, text=[], dpi=75, path=None):
    # draw image pair
    assert mkpts0.shape[0] == mkpts1.shape[0], f'mkpts0: {mkpts0.shape[0]} v.s. mkpts1: {mkpts1.shape[0]}'
    fig, axes = plt.subplots(1, 2, figsize=(10, 6), dpi=dpi)
    axes[0].imshow(img0, cmap='gray')
    axes[1].imshow(img1, cmap='gray')
    for i in range(2):   # clear all frames
        axes[i].get_yaxis().set_ticks([])
        axes[i].get_xaxis().set_ticks([])
        for spine in axes[i].spines.values():
            spine.set_visible(False)
    plt.tight_layout(pad=1)
    
    if kpts0 is not None:
        assert kpts1 is not None
        axes[0].scatter(kpts0[:, 0], kpts0[:, 1], c='w', s=2)
        axes[1].scatter(kpts1[:, 0], kpts1[:, 1], c='w', s=2)

    # draw matches
    if mkpts0.shape[0] != 0 and mkpts1.shape[0] != 0:
        fig.canvas.draw()
        transFigure = fig.transFigure.inverted()
        fkpts0 = transFigure.transform(axes[0].transData.transform(mkpts0))
        fkpts1 = transFigure.transform(axes[1].transData.transform(mkpts1))
        fig.lines = [matplotlib.lines.Line2D((fkpts0[i, 0], fkpts1[i, 0]),
                                            (fkpts0[i, 1], fkpts1[i, 1]),
                                            transform=fig.transFigure, c=color[i], linewidth=1)
                                        for i in range(len(mkpts0))]
        
        axes[0].scatter(mkpts0[:, 0], mkpts0[:, 1], c=color, s=4)
        axes[1].scatter(mkpts1[:, 0], mkpts1[:, 1], c=color, s=4)

    # put txts
    txt_color = 'k' if img0[:100, :200].mean() > 200 else 'w'
    fig.text(
        0.01, 0.99, '\n'.join(text), transform=fig.axes[0].transAxes,
        fontsize=15, va='top', ha='left', color=txt_color)

    # save or return figure
    if path:
        plt.savefig(str(path), bbox_inches='tight', pad_inches=0)
        plt.close()
    else:
        return fig
```

```python
# Define the dataset path
dataset_path = 'D:\\project\\tandt_db\\db\\playroom'  # Update with your dataset path
sparse_path = os.path.join(dataset_path, 'sparse\\0')

# Read cameras.bin file
cameras = read_write_model.read_cameras_binary(os.path.join(sparse_path, 'cameras.bin'))

# Print camera information
print("Cameras:")
for camera_id, camera in cameras.items():
    print(f"Camera ID: {camera_id}")
    print(f"Camera Model: {camera.model}")
    print(f"Width: {camera.width}")
    print(f"Height: {camera.height}")
    print(f"Parameters: {camera.params}")

# Read images.bin file
images = read_write_model.read_images_binary(os.path.join(sparse_path, 'images.bin'))

# Print image information and extrinsic parameters
print("\nImages:")
for image_id, image in images.items():
    print(f"Image ID: {image_id}")
    print(f"Camera ID: {image.camera_id}")
    print(f"Image Name: {image.name}")
    print(f"Rotation Vector (quaternion): {image.qvec}")
    print(f"Translation Vector: {image.tvec}")

    # Convert quaternion to rotation matrix
    rotation_matrix = read_write_model.qvec2rotmat(image.qvec)
    print(f"Rotation Matrix:\n{rotation_matrix}")

    # Camera position in world coordinates (inverse of extrinsic parameters)
    camera_position = -np.dot(rotation_matrix.T, image.tvec)
    print(f"Camera Position (world coordinates): {camera_position}")
    print()
```

```python
# Load example images
img0_pth = os.path.join(dataset_path, 'images\\DSC05573.jpg')
img1_pth = os.path.join(dataset_path, 'images\\DSC05575.jpg')
img0_raw = cv2.imread(img0_pth, cv2.IMREAD_GRAYSCALE)
img1_raw = cv2.imread(img1_pth, cv2.IMREAD_GRAYSCALE)
img0_raw = cv2.resize(img0_raw, (640, 480))
img1_raw = cv2.resize(img1_raw, (640, 480))

img0 = torch.from_numpy(img0_raw)[None][None].cuda() / 255.
img1 = torch.from_numpy(img1_raw)[None][None].cuda() / 255.
batch = {'image0': img0, 'image1': img1}

# Inference with LoFTR and get prediction
with torch.no_grad():
    out = loftr(batch)
    mkpts0 = out['keypoints0'].cpu().numpy()
    mkpts1 = out['keypoints1'].cpu().numpy()
    mconf = out['confidence'].cpu().numpy()

# Set the confidence threshold
confidence_threshold = 0.99

# Filter matches based on confidence
high_conf_indices = mconf > confidence_threshold
filtered_mkpts0 = mkpts0[high_conf_indices]
filtered_mkpts1 = mkpts1[high_conf_indices]
filtered_mconf = mconf[high_conf_indices]

# Draw
color = cm.jet(filtered_mconf)
text = [
    'LoFTR',
    'Matches: {}'.format(len(filtered_mkpts0)),
]
fig = make_matching_figure(img0_raw, img1_raw, filtered_mkpts0, filtered_mkpts1, color, text=text)
```

```python
import numpy as np

# 
fx = 1040.00730376
fy = 1040.19275667
cx = 632
cy = 416

# 
K = np.array([
    [fx, 0, cx],
    [0, fy, cy],
    [0, 0, 1]
])

# 
R = np.array([
    [0.99907574, 0.01713256, -0.03942251],
    [0.00643516, 0.84718388, 0.53126083],
    [0.04249997, -0.5310235, 0.84629061]
])

t = np.array([-1.39239957, 1.54667668, 2.24371909]).reshape(3, 1)

# 10
num_points = 10
image_width = 1264
image_height = 832

pixels = np.random.randint(0, min(image_width, image_height), size=(num_points, 2))
depths = np.random.uniform(0.5, 5.0, size=(num_points,))

# 
world_coordinates_list = []

for (pixel_x, pixel_y), depth in zip(pixels, depths):
    # 1
    normalized_x = (pixel_x - cx) / fx
    normalized_y = (pixel_y - cy) / fy

    # 2
    camera_coordinates = np.array([normalized_x * depth, normalized_y * depth, depth]).reshape(3, 1)

    # 3
    extrinsic_matrix = np.hstack((R, t))

    # 
    camera_coordinates_homogeneous = np.vstack((camera_coordinates, np.array([[1]])))
    print(camera_coordinates_homogeneous)

    # 
    world_coordinates_homogeneous = np.dot(extrinsic_matrix, camera_coordinates_homogeneous)

    # 
    world_coordinates = world_coordinates_homogeneous[:3].flatten()
    world_coordinates_list.append(world_coordinates)

    # 
    print(f"Pixel: ({pixel_x}, {pixel_y}), Depth: {depth:.2f}")
    print(f"3D world coordinates: {world_coordinates}\n")

# 
print("All computed 3D world coordinates:")
for i, world_coords in enumerate(world_coordinates_list):
    print(f"Point {i+1}: {world_coords}")

    
import numpy as np
import cv2

# 
camera_matrix = np.array([
    [fx, 0, cx],
    [0, fy, cy],
    [0, 0, 1]
], dtype=np.float32)

#  (distortion coefficients)
dist_coeffs = np.zeros((4, 1), dtype=np.float32)

# PnP
object_points = np.array(world_coordinates_list, dtype=np.float32)
image_points = np.array(pixels, dtype=np.float32)

success, rotation_vector, translation_vector = cv2.solvePnP(
    object_points, image_points, camera_matrix, dist_coeffs)

if success:
    print("Rotation Vector:\n", rotation_vector)
    print("Translation Vector:\n", translation_vector)

    # 
    rotation_matrix, _ = cv2.Rodrigues(rotation_vector)
    print("Rotation Matrix:\n", rotation_matrix)

    # 
    extrinsic_matrix_recomputed = np.hstack((rotation_matrix, translation_vector))
    print("Recomputed Extrinsic Matrix:\n", extrinsic_matrix_recomputed)

    # 
    camera_position_recomputed = -np.dot(rotation_matrix.T, translation_vector)
    print("Recomputed Camera Position (in world coordinates):\n", camera_position_recomputed)
else:
    print("PnP")
```
